```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
```

```python
#data = pd.read_csv('web_traffic.tsv', sep = '\t', names = ['1', '0'])
data = sp.genfromtxt('web_traffic.tsv', delimiter = '\t')
```

```python
data = pd.DataFrame(data)
data.head()
```

```python
data = data[~(sp.isnan(data['1']) | sp.isnan(data['0']))]
```

```python
x = data['0']
y = data['1']
```

```python
plt.scatter(x, y, s = 12)
plt.title("Web Traffic Over last month (original Data)")
plt.xlabel("Time")
plt.ylabel("Hits/hour")
plt.xticks([w*7*24 for w in range(6)], ['week %i' %w for w in range(6)])
plt.grid()
plt.show()
```

```python
plt.scatter(x, 0 * y, c = 'purple', s = 10)
plt.title("Data projected on X-axis")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
plt.scatter(0 * x, y, c = 'red', s = 8)
plt.title("Data projected on Y-axis")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
theta = 45
rot_matrix = [[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]]
data1 = data.dot(rot_matrix)
x = data1[0]
y = data1[1]
plt.scatter(x, y, s = 10, c = 'blue')
plt.title("Data represented in the xy-plane rotated counterclockwise through an angle of 45 degrees")
plt.grid()
plt.show()
```

```python
v = np.array([[1,1]]) # Vector Coordinates
data1 = np.matmul(data, v.T.dot(v) / (v.dot(v.T)))
x = data1[:, 0]
y = data1[:, 1]
plt.scatter(x, y, s = 12)
plt.title("Data projected on the x = y line")
plt.grid()
plt.show()
```
